```python
import pandas as pd

df = pd.read_csv("../files/cleaned_data.csv")
```

```python
plt.plot(df['Total'])
```

```python
avg_prices = df.groupby('Product')['Price'].mean()

# Plot histogram/bar chart
plt.figure(figsize=(12,6))
plt.bar(avg_prices.index, avg_prices.values)

plt.xlabel('Products')
plt.ylabel('Average Price')
plt.title('Average Price of Products')

plt.xticks(rotation=45)
plt.show()
```

```python
avg_prices = df.groupby('Product')['Price'].mean()

plt.figure(figsize=(12,6))
plt.bar(avg_prices.index, avg_prices.values)

plt.xlabel('Products')
plt.ylabel('Average Price')
plt.title('Average Product Prices')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Product'], df['Price'])

plt.xlabel('Products')
plt.ylabel('Price')
plt.title('Product Price Trend')

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['Price'], bins=20)

plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')

plt.show()
```

```python
product_sum = df.groupby('Product')['Price'].sum()

plt.figure(figsize=(8,8))
plt.pie(product_sum.values, labels=product_sum.index, autopct='%1.1f%%')

plt.title('Price Contribution by Product')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(df['Price'])

plt.ylabel('Price')
plt.title('Price Outliers')

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Quantity'], df['Price'])

plt.xlabel('Quantity')
plt.ylabel('Price')
plt.title('Quantity vs Price')

plt.show()
```

```python
avg_prices = df.groupby('Product')['Price'].mean()

plt.figure(figsize=(12,8))
plt.barh(avg_prices.index, avg_prices.values)

plt.xlabel('Average Price')
plt.ylabel('Products')
plt.title('Average Prices')

plt.show()
```

```python
df.groupby('Product')['Price'].mean().plot(kind='area')

plt.title('Area Chart of Product Prices')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.stem(df['Price'])

plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Stem Plot of Prices')

plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.step(range(len(df)), df['Price'])

plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Step Plot of Prices')

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.violinplot(df['Price'])

plt.ylabel('Price')
plt.title('Violin Plot')

plt.show()
```

```python
df['CumulativePrice'] = df['Price'].cumsum()

plt.figure(figsize=(12,6))
plt.plot(df['CumulativePrice'])

plt.xlabel('Index')
plt.ylabel('Cumulative Price')
plt.title('Cumulative Price Growth')

plt.show()
```

```python
df.boxplot(column='Price', by='Product', figsize=(12,6))

plt.xticks(rotation=45)
plt.title('Price Comparison by Product')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(range(len(df)), df['Price'])

plt.xlabel('Record Number')
plt.ylabel('Price')
plt.title('Price Scatter')

plt.show()
```

```python
sorted_prices = df['Price'].sort_values().reset_index(drop=True)

plt.figure(figsize=(12,6))
plt.plot(sorted_prices)

plt.xlabel('Sorted Index')
plt.ylabel('Price')
plt.title('Sorted Price Plot')

plt.show()
```

```python
stats = df.groupby('Product')['Price'].agg(['mean', 'std'])

plt.figure(figsize=(12,6))
plt.errorbar(stats.index, stats['mean'], yerr=stats['std'], fmt='o')

plt.xticks(rotation=45)
plt.ylabel('Price')
plt.title('Mean Price with Error Bars')

plt.show()
```

```python
plt.stackplot(range(len(df)), df['Price'])

plt.title('Stack Plot')
plt.show()
```

```python
import numpy as np

avg_prices = df.groupby('Product')['Price'].mean()
angles = np.linspace(0, 2*np.pi, len(avg_prices), endpoint=False)

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)
ax.plot(angles, avg_prices.values)

plt.title('Polar Price Chart')
plt.show()
```
