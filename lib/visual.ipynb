```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Trend")
plt.show()
```

```python
ages = [18, 19, 20, 21, 22, 18, 19, 20]

plt.hist(ages)
plt.title("Age Distribution")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("../files/cleaned_data.csv")
```

```python
plt.plot(df['Total'])
```

```python
avg_prices = df.groupby('Product')['Price'].mean()

# Plot histogram/bar chart
plt.figure(figsize=(12,6))
plt.bar(avg_prices.index, avg_prices.values)

plt.xlabel('Products')
plt.ylabel('Average Price')
plt.title('Average Price of Products')

plt.xticks(rotation=45)
plt.show()
```
