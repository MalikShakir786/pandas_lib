```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Sample dataset
df = pd.DataFrame({
    "Age": [22, 25, 30, 35, 40],
    "Salary": [30000, 35000, 50000, 65000, 80000]
})

# Select features
X = df[["Age", "Salary"]]

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled)
```
