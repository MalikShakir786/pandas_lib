```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]

print(outliers)
```

```python
outlier_products = outliers['Product'].unique()
outlier_products
```

```python
for product in outlier_products:
    mean_price = df[df['Product'] == product]['Price'].median()
    
    df.loc[
        (df['Product'] == product) &
        ((df['Price'] < lower_bound) | (df['Price'] > upper_bound)),
        'Price'
    ] = mean_price
```
